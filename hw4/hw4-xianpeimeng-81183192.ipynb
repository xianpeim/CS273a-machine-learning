```python
%matplotlib notebook
#%matplotlib inline
#used to plot in jupyter

import mltools as ml
import sys
import numpy as np
import matplotlib.pyplot as plt
#np.seterr(divide='ignore', invalid='ignore')

def printInfos(X, Y, round_para = 2):
    for i in range(X.shape[1]):
        tmp_X = X[:,i:i+1]
        print('X feature ', i, ' -> min:', int(min(tmp_X)),' max:', int(max(tmp_X))
              ,' mean:', round(np.mean(tmp_X), round_para),' var:', round(np.var(tmp_X),round_para))
    print("Y -> min:", min(Y), " X max:", max(Y),' mean:',
          round(np.mean(Y),round_para),' var:', round(np.var(Y),round_para))

# Data Loading
# 1.1 begin
X = np.genfromtxt('data/X_train.txt', delimiter=None)
Y = np.genfromtxt('data/Y_train.txt', delimiter=None)
X,Y = ml.shuffleData(X,Y)
print('-----Part 1.1:-----')
#printInfos(X,Y)
# 1.1 done
# 1.2 begin
Xtr, Xva, Ytr, Yva = ml.splitData(X, Y)
Xt, Yt = Xtr[:5000], Ytr[:5000] # subsample for efficiency (you can go higher)
XtS, params = ml.rescale(Xt) # Normalize the features
XvS, _ = ml.rescale(Xva, params) # Normalize the features
print('-----Part 1.2:-----')
print('Training:')
printInfos(XtS,Yt)
print('Validataion:')
printInfos(XvS,Yva)
# 1.2 done
# Linear Classifier
# 2.1 begin
print('-----Part 2.1:-----')
regs = []
tr_aucs = []
va_aucs = []

for i in range(10):
    reg = i*0.01
    regs.append(reg)
    learner = ml.linearC.linearClassify()
    learner.train(XtS, Yt, reg = reg, initStep=0.5, stopTol=1e-6, stopIter=100)
    tr_auc = learner.auc(XtS, Yt)
    va_auc = learner.auc(XvS, Yva)
    print(tr_auc, va_auc)
    tr_aucs.append(tr_auc) # training AUC
    va_aucs.append(va_auc) # validation AUC
plt.figure('2.1 Training and Validation AUC over reg')
plt.title('2.1 Training and Validation AUC over reg')
plt.plot(regs, tr_aucs, color='r', marker='o',label='Training')
plt.plot(regs, va_aucs, color='b', marker='o',label='Validation')
plt.legend()

# 2.1 done
# 2.2 begin
print('-----Part 2.2:-----')

XtS_ploy = ml.transforms.fpoly(XtS,2)
XvS_ploy = ml.transforms.fpoly(XvS,2)
print(XtS_ploy.shape, XvS_ploy.shape)
# 2.2 done
# 2.3 begin
print('-----Part 2.3:-----')
regs.clear()
tr_aucs.clear()
va_aucs.clear()
for i in range(5):
    reg = i*0.01
    regs.append(reg)
    learner = ml.linearC.linearClassify()
    learner.train(XtS_ploy, Yt, reg = reg, initStep=0.5, stopTol=1e-6, stopIter=100)
    tr_auc = learner.auc(XtS_ploy, Yt)
    va_auc = learner.auc(XvS_ploy, Yva)
    print(tr_auc, va_auc)
    tr_aucs.append(tr_auc) # training AUC
    va_aucs.append(va_auc) # validation AUC
plt.figure('2.3 Training and Validation AUC over reg on 2-poly data')
plt.title('2.3 Training and Validation AUC over reg on 2-poly data')
plt.plot(regs, tr_aucs, color='r', marker='o',label='Training')
plt.plot(regs, va_aucs, color='b', marker='o',label='Validation')
plt.legend()

# 3.1 begin
print('-----Part 3.1:-----')
Ks = []
tr_aucs = []
va_aucs = []

for i in range(5):
    K = i+1
    Ks.append(K)
    learner = ml.knn.knnClassify()
    learner.train(XtS, Yt, K = K, alpha=0.5)
    tr_auc = learner.auc(XtS, Yt)
    va_auc = learner.auc(XvS, Yva)
    print(tr_auc, va_auc)
    tr_aucs.append(tr_auc) # training AUC
    va_aucs.append(va_auc) # validation AUC
plt.figure('3.1 Training and Validation AUC over K')
plt.title('3.1 Training and Validation AUC over K')
plt.plot(Ks, tr_aucs, color='r', marker='o',label='Training')
plt.plot(Ks, va_aucs, color='b', marker='o',label='Validation')
plt.legend()
# 3.1 done
# 3.2 begin
print('-----Part 3.2:-----')
Ks.clear()
tr_aucs.clear()
va_aucs.clear()
for i in range(5):
    K = i+1
    Ks.append(K)
    learner = ml.knn.knnClassify()
    learner.train(Xt, Yt, K = K, alpha=0.5)
    tr_auc = learner.auc(Xt, Yt)
    va_auc = learner.auc(Xva, Yva)
    print(tr_auc, va_auc)
    tr_aucs.append(tr_auc) # training AUC
    va_aucs.append(va_auc) # validation AUC
plt.figure('3.2 Training and Validation AUC over K on unscaled data')
plt.title('3.2 Training and Validation AUC over K on unscaled data')
plt.plot(Ks, tr_aucs, color='r', marker='o',label='Training')
plt.plot(Ks, va_aucs, color='b', marker='o',label='Validation')
plt.legend()
# 3.2 done

# 3.3 begin
print('-----Part 3.3:-----')

K = range(1,6,1)
A = range(0,3,1)
tr_aucs = np.zeros((len(K),len(A)))
va_aucs = np.zeros((len(K),len(A)))
for i,k in enumerate(K):
    for j,a in enumerate(A):
        learner = ml.knn.knnClassify()
        learner.train(XtS, Yt, K = k, alpha=a/2)
        tr_auc = learner.auc(XtS, Yt)
        va_auc = learner.auc(XvS, Yva)
        print(tr_auc, va_auc)
        tr_aucs[i][j] = tr_auc
        va_aucs[i][j] = va_auc
        
f, ax = plt.subplots(1, 2)
cax = ax[0].matshow(tr_aucs, interpolation='nearest')
f.colorbar(cax, ax=ax[0])
ax[0].title.set_text('Training AUC')
ax[0].set_xticklabels(A)
ax[0].set_yticklabels(K)
cax1 = ax[1].matshow(va_aucs, interpolation='nearest')
f.colorbar(cax1, ax=ax[1])
ax[1].title.set_text('Validation AUC')
ax[1].set_xticklabels(A)
ax[1].set_yticklabels(K)

# 3.3 done
# 4.1 begin

print('-----Part 4.1:-----')
maxDs = []
tr_aucs = []
va_aucs = []
for i in range(15):
    maxD = i
    maxDs.append(maxD)
    learner = ml.dtree.treeClassify(XtS, Yt, maxDepth=maxD, minParent=2,minLeaf=1)
    tr_auc = learner.auc(XtS, Yt)
    va_auc = learner.auc(XvS, Yva)
    print(tr_auc, va_auc)
    tr_aucs.append(tr_auc) # training AUC
    va_aucs.append(va_auc) # validation AUC
plt.figure('4.1 Training and Validation AUC over maxDepth')
plt.title('4.1 Training and Validation AUC over maxDepth')
plt.plot(maxDs, tr_aucs, color='r', marker='o',label='Training')
plt.plot(maxDs, va_aucs, color='b', marker='o',label='Validation')
plt.legend()
# 4.1 done
# 4.2 begin
print('-----Part 4.2:-----')
maxDs = []
nums = []
nums1 = []
for i in range(15):
    maxD = i
    maxDs.append(maxD)
    learner = ml.dtree.treeClassify(XtS, Yt, maxDepth=maxD, minParent=2,minLeaf=1)
    num = learner.sz
    learner = ml.dtree.treeClassify(XtS, Yt, maxDepth=maxD, minParent=4,minLeaf=1)
    num1 = learner.sz
    print(num, num1)
    nums.append(num) # training AUC
    nums1.append(num1) # validation AUC
plt.figure('4.2 Tree node number over maxDepth and minParent')
plt.title('4.2 Tree node number over maxDepth and minParent')
plt.plot(maxDs, nums, color='r', marker='o',label='minParent = 2')
plt.plot(maxDs, nums1, color='b', marker='o',label='minParent = 4')
plt.legend()
# 4.2 done
# 4.3 begin
print('-----Part 4.3:-----')
maxP = range(1,10,1)
minL = range(1,10,1)
tr_aucs = np.zeros((len(maxP),len(minL)))
va_aucs = np.zeros((len(maxP),len(minL)))
for i,maxp in enumerate(maxP):
    for j,minl in enumerate(minL):
        learner = ml.dtree.treeClassify(XtS, Yt, maxDepth=5, minParent=maxp, minLeaf=minl)
        tr_auc = learner.auc(XtS, Yt)
        va_auc = learner.auc(XvS, Yva)
        #print(tr_auc, va_auc)
        tr_aucs[i][j] = tr_auc
        va_aucs[i][j] = va_auc
        
f, ax = plt.subplots(1, 2)
cax = ax[0].matshow(tr_aucs, interpolation='nearest')
f.colorbar(cax, ax=ax[0])
ax[0].title.set_text('Training AUC')
ax[0].set_xticklabels(minL)
ax[0].set_yticklabels(maxP)
cax1 = ax[1].matshow(va_aucs, interpolation='nearest')
f.colorbar(cax1, ax=ax[1])
ax[1].title.set_text('Validation AUC')
ax[1].set_xticklabels(minL)
ax[1].set_yticklabels(maxP)
# 4.3 done
# 5.1 begin
print('-----Part 5.1:-----')
layers = range(3,6,1)
nodes = range(4,7,1)
tr_aucs = np.zeros((len(layers),len(nodes)))
va_aucs = np.zeros((len(layers),len(nodes)))
for i,layer in enumerate(layers):
    for j,node in enumerate(nodes):
        size = [XtS.shape[1]]
        for z in range(layer):
            size.append(node)
        size.append(2)
        nn = ml.nnet.nnetClassify()
        nn.init_weights(size, 'random', XtS, Yt)
        nn.train(XtS, Yt, stopTol=1e-8, stepsize=.25, stopIter=300)
        tr_auc = nn.auc(XtS, Yt)
        va_auc = nn.auc(XvS, Yva)
        print(tr_auc, va_auc)
        tr_aucs[i][j] = tr_auc
        va_aucs[i][j] = va_auc
        
f, ax = plt.subplots(1, 2)
cax = ax[0].matshow(tr_aucs, interpolation='nearest')
f.colorbar(cax, ax=ax[0])
ax[0].title.set_text('Training AUC')
ax[0].set_xticklabels(nodes)
ax[0].set_yticklabels(layers)
cax1 = ax[1].matshow(va_aucs, interpolation='nearest')
f.colorbar(cax1, ax=ax[1])
ax[1].title.set_text('Validation AUC')
ax[1].set_xticklabels(nodes)
ax[1].set_yticklabels(layers)

# 5.2 begin
print('-----Part 5.2:-----')
custs = [[],[]]
logs = [[],[]]
htangs = [[],[]]
layers = 5;
nodes = range(3,8,1)
for i,node in enumerate(nodes):
    size = [XtS.shape[1]]
    for z in range(layers):
        size.append(node)
    size.append(2)
    #cust
    nn = ml.nnet.nnetClassify()
    nn.init_weights(size, 'random', XtS, Yt)
    def sig(z): return np.atleast_2d(z)
    def dsig(z): return np.atleast_2d(1)
    nn.setActivation('custom', sig, dsig)
    nn.train(XtS, Yt, stopTol=1e-8, stepsize=.25, stopIter=300)
    tr_auc = nn.auc(XtS, Yt)
    va_auc = nn.auc(XvS, Yva)
    print(tr_auc, tr_auc)
    custs[0].append(tr_auc)
    custs[1].append(va_auc)
    #log
    nn = ml.nnet.nnetClassify()
    nn.init_weights(size, 'random', XtS, Yt)
    nn.setActivation('logistic')
    nn.train(XtS, Yt, stopTol=1e-8, stepsize=.25, stopIter=300)
    tr_auc = nn.auc(XtS, Yt)
    va_auc = nn.auc(XvS, Yva)
    print(tr_auc, tr_auc)
    logs[0].append(tr_auc)
    logs[1].append(va_auc)
    #htang
    nn = ml.nnet.nnetClassify()
    nn.init_weights(size, 'random', XtS, Yt)
    nn.setActivation('htangent')
    nn.train(XtS, Yt, stopTol=1e-8, stepsize=.25, stopIter=300)
    tr_auc = nn.auc(XtS, Yt)
    va_auc = nn.auc(XvS, Yva)
    print(tr_auc, tr_auc)
    htangs[0].append(tr_auc)
    htangs[1].append(va_auc)
    
plt.figure('5.2 Training and Validation AUC over node number with custom activation')
plt.title('5.2 Training and Validation AUC over node number with custom activation')
plt.plot(list(nodes), custs[0], color='r', marker='o',label='Training')
plt.plot(list(nodes), custs[1], color='b', marker='o',label='Validation')
plt.legend()
plt.figure('5.2 Training and Validation AUC over node number with logistic activation')
plt.title('5.2 Training and Validation AUC over node number with logistic activation')
plt.plot(list(nodes), logs[0], color='r', marker='o',label='Training')
plt.plot(list(nodes), logs[1], color='b', marker='o',label='Validation')
plt.legend()
plt.figure('5.2 Training and Validation AUC over node number with htangent activation')
plt.title('5.2 Training and Validation AUC over node number with htangent activation')
plt.plot(list(nodes), htangs[0], color='r', marker='o',label='Training')
plt.plot(list(nodes), htangs[1], color='b', marker='o',label='Validation')
plt.legend()
# 5.2 done
```
